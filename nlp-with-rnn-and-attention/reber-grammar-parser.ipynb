```python
import random

from embedded_reber_grammar import EmbeddedReberGrammar, ReberGrammar

grammar = EmbeddedReberGrammar()
is_valid_grammar_set = [','.join(grammar.produce()) for x in range(50_000)]

for i in is_valid_grammar_set:
    print(i)

max_length = max(map(len, is_valid_grammar_set), default=2)
print(max_length)
```

```python
letters = ReberGrammar().get_edge_labels()
is_invalid_grammar_set = []

max_length = max(map(len, is_valid_grammar_set), default=2)
print(max_length)
for i in range(50_000):
    invalid_grammar_string  = ""
    for i in range(random.randint(1, max_length)):
        invalid_grammar_string += random.choice(letters) + ','
    is_invalid_grammar_set.append(invalid_grammar_string)

is_invalid_grammar_set
```
