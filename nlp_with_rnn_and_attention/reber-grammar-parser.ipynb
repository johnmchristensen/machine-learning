```python
from reber_dataset import ReberDataset
from torch.utils.data import DataLoader

train_ds = ReberDataset(10_000)
train_loader = DataLoader(train_ds, batch_size=32, shuffle=True)

valid_ds = ReberDataset(10_000)
valid_loader = DataLoader(valid_ds, batch_size=32, shuffle=True)
```

```python
from model_runner.ModelRunner import ModelRunner
from reber_detector import EmbeddedReberDetector
from torch import nn
from torchmetrics.classification import BinaryF1Score

model = EmbeddedReberDetector()
runner = ModelRunner(model, BinaryF1Score(threshold=0.5), loss_fn=nn.BCEWithLogitsLoss(), device="cpu")
runner.train_model(train_loader, valid_loader, patience=5)
```

```python
from torchmetrics.classification import BinaryAccuracy

# Fresh, unseen data for testing. train_model already reloaded the best checkpoint
# into runner.model, so this evaluates the best model.
test_ds = ReberDataset(10_000)
test_loader = DataLoader(test_ds, batch_size=32)

accuracy = runner.test_model(test_loader, metric=BinaryAccuracy(threshold=0.5))
print(f"Test accuracy: {accuracy.item():.4f}")
```
