```python
from reber_dataset import ReberDataset
from torch.utils.data import DataLoader

train_ds = ReberDataset(10_000)
train_loader = DataLoader(train_ds, batch_size=32, shuffle=True)

valid_ds = ReberDataset(10_000)
valid_loader = DataLoader(valid_ds, batch_size=32, shuffle=True)
```

```python
from model_runner.ModelRunner import ModelRunner
from reber_detector import EmbeddedReberDetector
from torch import nn
from torchmetrics.classification import BinaryF1Score

model = EmbeddedReberDetector()
runner = ModelRunner(model, BinaryF1Score(threshold=0.5), loss_fn=nn.BCEWithLogitsLoss())
runner.train_model(train_loader, valid_loader, patience=5)
```
